REVENUE ANALYSIS

```python
#What country generates the most revenue?
```

```python
df=pd.read_sql_query("""SELECT BillingCountry AS Countries, SUM(Total) AS Total_Revenue FROM invoices GROUP BY Countries ORDER BY Total_revenue DESC LIMIT 1;""",conn)
df
```

```python
#What city generates the most revenue?
df=pd.read_sql_query("""SELECT BillingCity AS City, SUM(Total) AS Total_Revenue FROM invoices GROUP BY City ORDER BY Total_revenue DESC LIMIT 1;""",conn)
df
```

```python
#What is the average invoice value per country
df=pd.read_sql_query("""SELECT BillingCountry AS Country, AVG(Total)AS Average_Invoice_per_country  FROM invoices GROUP BY Country ORDER BY Average_Invoice_per_country DESC;""",conn)
df
```

```python
#Which year had the Highest revenue?
df=pd.read_sql_query("""SELECT strftime('%Y',InvoiceDate) AS YEAR, SUM(Total)AS  Revenue  FROM invoices GROUP BY YEAR  ORDER BY Revenue DESC LIMIT 1 ;""",conn)
df
```

```python
#What are the Top 5 Revenue generting countries?
df=pd.read_sql_query("""SELECT BillingCountry AS Country, SUM(Total) AS Total_Revenue FROM invoices GROUP BY Country ORDER BY Total_Revenue DESC LIMIT 5;""",conn)
df
```

CUSTOMER BEHAVIOUR

```python
#Who are the TOP 10 Customers by Total Spending?
df=pd.read_sql_query("""SELECT C.FirstName AS First_Name, C.LastName AS Last_Name, SUM(I.Total) AS Total  FROM customers AS C JOIN invoices AS I ON C.CustomerId= I.CustomerId GROUP BY First_Name, Last_Name  ORDER BY Total DESC LIMIT 10;""",conn)
df
```

```python
#Which countries have the most customers?
df=pd.read_sql_query("""SELECT Country, COUNT(CustomerId) AS Number_of_customers FROM customers GROUP BY Country ORDER BY Number_of_customers DESC;""",conn)
df
```

```python
#Do customers from some countries spend more than others?
df=pd.read_sql_query("""SELECT  C.Country AS Country, C.FirstName AS First_Name, C.LastName AS Last_Name, SUM(I.Total) AS Amount_Spent FROM customers AS C JOIN Invoices AS I ON C.CustomerId=I.CustomerId GROUP BY Country, First_Name, Last_Name ORDER BY Amount_Spent  DESC LIMIT 10;""",conn)
df
```

```python
#What is the average amount spent by the customer?
df=pd.read_sql_query("""SELECT C.FirstName AS First_Name, C.LastName AS Last_Name, AVG(I.Total) AS Amount_Spent FROM customers AS C JOIN Invoices AS I ON C.CustomerId=I.CustomerId  GROUP BY First_Name, Last_Name ORDER BY Amount_Spent  DESC LIMIT 10;""",conn)
df
```

PRODUCTS AND MUSIC TREND

```python
#Which Music genre sells the most?
df=pd.read_sql_query("""SELECT G.Name, COUNT(I.TrackId) AS TOTAL FROM invoice_items AS I JOIN Tracks AS T ON I.TrackId=T.TrackId JOIN genres AS G ON T.TrackId=G.GenreId GROUP BY G.Name ORDER BY TOTAL DESC;""",conn)
df
```

```python
#Who are the Top selling Artists?
df=pd.read_sql_query("""SELECT AR.Name AS Artists, COUNT(I.TrackId) AS Total_Revenue FROM Invoice_items I JOIN Tracks T ON I.TrackId=T.TrackId  JOIN Albums A ON T.AlbumId= A.AlbumId JOIN artists AR ON A.ArtistId=AR.ArtistId GROUP BY Artists ORDER BY Total_Revenue LIMIT 10;""",conn)
df
```

```python
#What is the Top selling Track?
df=pd.read_sql_query("""SELECT T.Name AS Tracks, COUNT(I.TrackId) AS Top_selling_Track FROM invoice_items AS I JOIN Tracks AS T ON I.TrackId=T.TrackId GROUP BY Tracks ORDER BY Top_selling_Track DESC LIMIT 1;""",conn)
df
```

```python
#Which genre generates the most revenue?
df=pd.read_sql_query("""SELECT G.Name AS Genre, SUM(I.UnitPrice*Quantity) AS Revenue  FROM invoice_items AS I JOIN Tracks AS T ON I.invoiceId=T.TrackId JOIN Genres AS G ON T.GenreId=G.GenreId GROUP BY Genre ORDER BY Revenue DESC  LIMIT 1;""",conn)
df
```

EMPLOYEE PERFORMANCE

```python
#which employee supports the highest spending customer?
df=pd.read_sql_query("""SELECT E.FirstName AS EMPLOYEE_FIRST_NAME, E.LastName AS EMPLOYEE_LAST_NAME, C.FirstName AS CUSTOMER_FIRST_NAME,C.LastName AS CUSTOMER_LAST_NAME, SUM(I.Total) AS AMOUNT_SPENT  FROM employees AS E JOIN customers AS C ON E.EmployeeId= C.SupportRepId JOIN Invoices AS I ON C.CustomerId=I.CustomerId GROUP BY EMPLOYEE_FIRST_NAME, EMPLOYEE_LAST_NAME, CUSTOMER_FIRST_NAME, CUSTOMER_LAST_NAME ORDER BY AMOUNT_SPENT DESC LIMIT 11;""",conn)
df
```

```python
#Which employee  generates the most revenue?
df=pd.read_sql_query("""SELECT E.FirstName AS First_Name, E.LastName AS Last_Name, SUM(I.Total) AS Total_Revenue FROM employees AS E JOIN customers C ON E.EmployeeId=C.SupportRepId  JOIN Invoices I ON C.CustomerId=I.CustomerId GROUP BY First_Name, Last_Name ORDER BY Total_Revenue DESC LIMIT 1;""",conn)
df
```
